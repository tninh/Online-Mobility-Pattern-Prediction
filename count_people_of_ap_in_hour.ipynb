```python
import pandas as pd
from datetime import datetime
import time

df = pd.read_csv('data_sample/ap.csv')
df = df[['Client Username', 'Association Time', 'AP Name']]
df
```

```python
count_dict = {}
time_list = df[df['AP Name'] == 'CVC-5-521-1(new)']
# print(time_list)
for t in time_list['Association Time']:
    t = datetime.strptime(t, '%a %b %d %H:%M:%S PST %Y')
    t_sec = int(time.mktime(t.timetuple()) / 3600)
    if t_sec in count_dict:
        count_dict[t_sec] += 1
    else:
        count_dict[t_sec] = 1
count_dict
```

```python
for key, value in count_dict.items():
    new_key = datetime.fromtimestamp(key * 3600).strftime('%a %b %d %H:%M:%S PST %Y')
    print(new_key, 'number of people:', value)
```
