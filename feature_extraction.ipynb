```python
def convertMonth(month):
    if month == 'Jan': return 1
    elif month == 'Feb': return 2
    elif month == 'Mar': return 3
    elif month == 'Apr': return 4
    elif month == 'May': return 5
    elif month == 'Jun': return 6
    elif month == 'Jul': return 7
    elif month == 'Aug': return 8
    elif month == 'Sep': return 9
    elif month == 'Oct': return 10
    elif month == 'Nov': return 11
    elif month == 'Dec': return 12
    else: print('error input is not correct')
        
def convert_time_slot(time):
    temp = (time.hour * 60 + time.minute) * 60 + time.second
    return int(temp/(5*60) + 1)
```

```python
data_dict = defaultdict(list)
for i in range(df.shape[0]):
    temp = df.iloc[i,1].split()
    temp1 = df.iloc[i,3].split()

    if len(temp1) == 2: sess_duration = int(temp1[0])
    elif len(temp1) == 4: sess_duration = int(temp1[0]) * 60 + int(temp1[2])
    elif len(temp1) == 6: sess_duration = (int(temp1[0]) * 60 + int(temp1[2])) * 60 + int(temp1[4])
    else: print('error')

    d_time = datetime.datetime(year=int(temp[5]), month=convertMonth(temp[1]), day=int(temp[2]), hour=int(temp[3].split(':')[0]), minute=int(temp[3].split(':')[1]), second=int(temp[3].split(':')[2]))
    data_dict[df.iloc[i,0]].append([d_time, df.iloc[i,2], sess_duration])
    
# for i in data_dict.items():
#     i[1].sort()
#     for j in reversed(range(1,len(i[1]))):
#         if i[1][j][1] == i[1][j-1][1]:
#             i[1][j-1][2] += i[1][j][2]
#             i[1][j][2] = 0
    
for i in data_dict:
    data_dict[i].sort()
    for j in reversed(range(1,len(data_dict[i]))):
        if data_dict[i][j][1] == data_dict[i][j-1][1]:
            data_dict[i][j-1][2] += data_dict[i][j][2]
            data_dict[i][j][2] = 0
    
    data_dict[i] = [x for x in data_dict[i] if x[2] > 0]
    for j in range(len(data_dict[i])):
        data_dict[i][j][2] = int(data_dict[i][j][2]/(5*60)) + 1
```
